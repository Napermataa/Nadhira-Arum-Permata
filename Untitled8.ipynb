```python
import pandas as pd
df = pd.read_excel('/content/drive/MyDrive/ECM.xlsx')
```

```python
print(df)
```

```python
df['Purchase_Amount'] = df['Purchase_Amount'].replace('[\$,]', '', regex=True).astype(float)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(df['Age'], bins=20, kde=True)
plt.title("Distribusi Umur Konsumen")
plt.show()
```

```python
category_counts = df['Purchase_Category'].value_counts()

# Mengambil 5 kategori dengan jumlah pembelian tertinggi
top_5_categories = category_counts.head(5)

# Menampilkan hasil
print("Top 5 Kategori dengan Jumlah Pembelian Tertinggi:")
print(top_5_categories)
```

```python
overall_median = df['Purchase_Amount'].median()

# Calculate median purchase amount for each category
median_price_per_category = df.groupby('Purchase_Category')['Purchase_Amount'].median()

# Apply the categorization based on the overall median
price_category = median_price_per_category.apply(lambda x: 'Tinggi' if x > overall_median else 'Rendah')

# Display the price categories
print("Kategori Harga untuk Top 5 Kategori:")
print(price_category)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Menghitung median harga pembelian per kategori
median_price_per_category = df.groupby('Purchase_Category')['Purchase_Amount'].median().sort_values(ascending=False)

# Membuat plot bar
plt.figure(figsize=(12, 6))  # Mengatur ukuran plot
sns.barplot(x=median_price_per_category.index, y=median_price_per_category.values)
plt.title('Median Harga Pembelian per Kategori')
plt.xlabel('Kategori Pembelian')
plt.ylabel('Median Harga Pembelian')
plt.xticks(rotation=45, ha='right')  # Rotasi label sumbu x agar mudah dibaca
plt.tight_layout()  # Menyesuaikan tata letak plot
plt.show()
```

```python
sns.boxplot(x='Income_Level', y='Purchase_Amount', data=df)
plt.title("Distribusi Pembelian berdasarkan Tingkat Pendapatan")
plt.show()


```

```python
# Get the top 5 categories
category_counts = df['Purchase_Category'].value_counts()
top_5_categories = category_counts.head(5).index  # Get the index (category names)

# Filter the original dataframe to include only the top 5 categories
df_top_5 = df[df['Purchase_Category'].isin(top_5_categories)]

# Now you can proceed with the rest of your analysis
top_5_rating_avg = df_top_5.groupby('Purchase_Category')['Product_Rating'].mean()

# Menampilkan hasil
print("Rata-rata Rating untuk Top 5 Kategori:")
print(top_5_rating_avg)

# Menentukan apakah rating tinggi atau tidak berdasarkan median keseluruhan
overall_rating_median = df['Product_Rating'].median()
rating_category = top_5_rating_avg.apply(lambda x: 'Tinggi' if x > overall_rating_median else 'Rendah')

# Menampilkan kategori rating
print("Kategori Rating untuk Top 5 Kategori:")
print(rating_category)

# Analisis apakah rating mempengaruhi banyaknya pembelian
correlation = df[['Product_Rating', 'Purchase_Amount']].corr().iloc[0, 1]
print("Korelasi antara Rating dan Jumlah Pembelian:", correlation)
```

```python
# Menghitung rata-rata umur pembeli untuk masing-masing kategori top 5
average_age_per_category = df_top_5.groupby('Purchase_Category')['Age'].mean()

# Menampilkan hasil
print("Rata-rata Umur Pembeli untuk Top 5 Kategori:")
print(average_age_per_category)
```

```python

numeric_df = df.select_dtypes(include=np.number)
corr = numeric_df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matriks Korelasi Variabel Numerik")
plt.show()
```
